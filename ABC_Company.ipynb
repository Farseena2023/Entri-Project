1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python

#Calculate number of employees in each team
team_employees=ABC_company["Team"].value_counts()
print("Number of employees in each team:")
team_employees
```

```python
# Calculate the percentage splitting with respect to the total employees
percentage=(team_employees/len(ABC_company))*100
print("\nPercentage splitting with respect to total employees:")
percentage
```

3.Find from which age group most of the employees belong to

```python

age=ABC_company['Age'].value_counts().head()
age
```

```python
#Age group distribution
age_groups = pd.cut(ABC_company['Age'], 
                    bins=[20, 25, 30, 35, 40],
                    labels=['20-25','25-30', '30-35', '35-40'])
age_group_counts = age_groups.value_counts()
age_groups
```

```python
plt.hist(ABC_company['Age'])
plt.xlabel("age")
plt.ylabel("no. of employees")
plt.grid()
plt.show()
```

Employees are mostly of age group 23 to 25 years,most employees are of age 24

4.Find out under which team and position, spending in terms of salary is high

```python
# High spending team 
high_spending_team=ABC_company.groupby('Team')['Salary'].sum().idxmax()
high_spending_team
```

```python
# High spending position
high_spending_position=ABC_company.groupby('Position')['Salary'].sum().idxmax()
high_spending_position
```

```python
high_salary_team_position = ABC_company.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
high_salary_team_position
```

5.Find if there is any correlation between age and salary , represent it visually.

```python
#correlation between age and salary
correlation= ABC_company[['Age', 'Salary']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Age and Salary')
plt.show()
```
